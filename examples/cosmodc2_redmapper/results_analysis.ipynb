```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

DIR_BASE    = "/sps/lsst/users/ebarroso/CLPipeline/examples/cosmodc2_redmapper/baseline"
DIR_HMF     = "/sps/lsst/users/ebarroso/CLPipeline/examples/cosmodc2_redmapper/hmf_analysis"
DIR_EXCL    = "/sps/lsst/users/ebarroso/CLPipeline/examples/cosmodc2_redmapper/selection_analysis/exclude_first_rich_bin"
DIR_COSMO   = "/sps/lsst/users/ebarroso/CLPipeline/examples/cosmodc2_redmapper/deltasigma_cosmology_analysis"

HMFS     = ["bocquet16", "tinker08", "tinker10", "watson13"]
HMFS_ALL = HMFS + ["despali16"]

COSMO_POINTS = ["cosmo_0.22_0.7", "cosmo_0.3_0.9", "cosmo_0.4_0.8"]
COSMO_LABELS = [r"$\Omega_c=0.22,\ \sigma_8=0.7$",
                r"$\Omega_c=0.3,\ \sigma_8=0.9$",
                r"$\Omega_c=0.4,\ \sigma_8=0.8$"]

COLORS       = ["#000000", "#e41a1c", "#377eb8", "#4daf4a", "#984ea3"]
COLORS_2     = ["#e41a1c", "#000000"]

PARAMS_COSMO = [
    (r"$\Omega_c$", 0.22),
    (r"$\sigma_8$", 0.80),
]
PARAMS_MOR = [
    (r"$\mu_0$",     3.34),
    (r"$\mu_m$",     0.958236982),
    (r"$\mu_z$",    -0.0192802),
    (r"$\sigma_0$",  0.562317194),
    (r"$\sigma_m$",  0.04552506),
    (r"$\sigma_z$",  -0.0445),
]
PARAMS_BOTH = PARAMS_COSMO + PARAMS_MOR


def _samples_path(base, run):
    return f"{base}/run_in2p3_{run}/outputs_{run}/output_rp/number_counts_samples.txt"

def hmf_path(name, run):
    return _samples_path(f"{DIR_HMF}/{name}", run)

def baseline_path(analysis, run):
    return _samples_path(f"{DIR_BASE}/{analysis}", run)

def excl_path(run):
    return f"{DIR_EXCL}/run_in2p3_{run}/outputs_{run}/output_rp/number_counts_samples.txt"

def cosmo_path(point, run, lensing=False):
    folder = f"{point}_lensing" if lensing else point
    return _samples_path(f"{DIR_COSMO}/{folder}", run)


def build_hmf_paths(run):
    return [hmf_path(n, run) for n in HMFS] + [baseline_path("cosmodc2_redmapper_full_analysis", run)]

def build_excl_paths(run):
    return [excl_path(run), baseline_path("cosmodc2_redmapper_full_analysis", run)]

def build_cosmo_paths(run, lensing=False):
    return [cosmo_path(p, run, lensing) for p in COSMO_POINTS]


def load_chain(path, param_names):
    try:
        return _load_cosmosis_chain(path, param_names, 0.3)
    except Exception:
        from getdist import MCSamples
        data = np.loadtxt(path)
        if data.shape[0] < data.shape[1]:
            data = data.T
        return MCSamples(samples=data, names=param_names, labels=param_names)


def summarize(samples, param_names):
    mean   = np.array([np.median(samples.samples[:, samples.index[p]]) for p in param_names])
    std    = np.array([samples.std(p) for p in param_names])
    return mean, std


def plot_sensitivity(paths, labels, params, colors, title=""):
    param_labels = [p for p, _ in params]
    fiducials    = np.array([v for _, v in params])
    n_params     = len(params)
    n_models     = len(paths)

    all_means, all_stds = [], []
    for path in paths:
        samples   = load_chain(path, param_labels)
        mean, std = summarize(samples, param_labels)
        all_means.append(mean)
        all_stds.append(std)

    all_means = np.array(all_means)
    all_stds  = np.array(all_stds)

    fig, axes = plt.subplots(
        1, n_params,
        #figsize=(4.5 * n_params, n_models * 0.8 + 2.0),
        figsize=(10, 6),
        sharey=True,
    )
    if n_params == 1:
        axes = [axes]

    y_pos = np.arange(n_models)[::-1]

    for p_idx, (ax, plabel, fid) in enumerate(zip(axes, param_labels, fiducials)):
        ax.axvline(fid, color="gray", linewidth=1.2, linestyle="--", alpha=0.7, zorder=0)

        for y, mean, std, color in zip(y_pos, all_means, all_stds, colors):
            ax.errorbar(
                mean[p_idx], y,
                xerr=std[p_idx],
                fmt="o", color=color,
                capsize=3, capthick=1,
                linewidth=1.2, markersize=5,
                zorder=2,
            )

        ax.set_xlabel(plabel, fontsize=12)
        ax.xaxis.set_major_locator(ticker.MaxNLocator(4))
        ax.tick_params(axis="x", labelsize=8)
        ax.grid(axis="x", alpha=0.25, linestyle=":")
        ax.spines[["top", "right"]].set_visible(False)

        if p_idx == 0:
            ax.set_yticks(y_pos)
            ax.set_yticklabels(labels, fontsize=10)
            ax.spines["left"].set_visible(False)
            ax.tick_params(left=False)
        else:
            ax.spines["left"].set_alpha(0.2)

    fig.subplots_adjust(wspace=0.08)
    if title:
        fig.suptitle(title, fontsize=12, y=1.01)
    plt.tight_layout()
    plt.show()


# ── HMF sensitivity ───────────────────────────────────────────────────────────

plot_sensitivity(build_hmf_paths("mor"),  HMFS_ALL, PARAMS_MOR,   COLORS, title="HMF sensitivity — MOR")
plot_sensitivity(build_hmf_paths("both"), HMFS_ALL, PARAMS_BOTH,  COLORS, title="HMF sensitivity — BOTH")
#plot_sensitivity(build_hmf_paths("cosmo"),HMFS_ALL, PARAMS_COSMO, COLORS, title="HMF sensitivity — COSMO")

# ── Selection: exclude first richness bin ─────────────────────────────────────
#plot_sensitivity(build_excl_paths("mor"),  ["exclude first bin", "baseline"], PARAMS_MOR,  COLORS_2, title="Selection — MOR")
plot_sensitivity(build_excl_paths("both"), ["exclude first bin", "baseline"], PARAMS_BOTH, COLORS_2, title="Selection — BOTH")

# ── Delta Sigma Cosmology: full analysis (counts + lensing) ───────────────────────────────

plot_sensitivity(build_cosmo_paths("both", lensing=False), COSMO_LABELS, PARAMS_BOTH,  COLORS[:3], title="Delta Sigma Cosmology: full analysis")

# ── Delta Sigma Cosmology: lensing only vs lensing baseline ───────────────────────────────

lensing_paths  = build_cosmo_paths("both", lensing=True) + [baseline_path("cosmodc2_redmapper_lensing", "both")]
lensing_labels = COSMO_LABELS + ["lensing baseline"]

plot_sensitivity(lensing_paths, lensing_labels, PARAMS_BOTH, COLORS[:4], title="Delta Sigma Cosmology: lensing only")
```

```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def plot_deviation_heatmap(paths, labels, params, title=""):

    param_labels = [p for p, _ in params]
    fiducials    = np.array([v for _, v in params])

    n_models = len(paths)
    n_params = len(params)

    all_medians, all_stds = [], []

    for path in paths:
        samples = load_chain(path, param_labels)
        median, std = summarize(samples, param_labels)

        all_medians.append(median)
        all_stds.append(std)

    all_medians = np.array(all_medians)
    all_stds    = np.array(all_stds)

    # ─────────────────────────────────────────────
    # RELATIVE DEVIATION (THIS IS WHAT YOU WANT)
    # ─────────────────────────────────────────────
    eps = 1e-8
    
    rel_dev = (all_medians - fiducials) / (np.abs(all_medians) + np.abs(fiducials) + eps)

    # optional diagnostic (keep if needed)
    pull = (all_medians - fiducials) / all_stds

    # ─────────────────────────────────────────────
    # PLOT
    # ─────────────────────────────────────────────
    fig, ax = plt.subplots(
        figsize=(max(8, n_params * 1.1), max(3, n_models * 0.7))
    )

    cmap = mcolors.LinearSegmentedColormap.from_list(
        "rw_b", ["#c0392b", "white", "#2980b9"]
    )

    vmax = np.nanmax(np.abs(rel_dev))

    im = ax.imshow(
        rel_dev,
        cmap=cmap,
        vmin=-vmax,
        vmax=vmax,
        aspect="auto"
    )

    # ─────────────────────────────────────────────
    # TEXT = %
    # ─────────────────────────────────────────────
    for i in range(n_models):
        for j in range(n_params):
            pct = rel_dev[i, j] * 100
            ax.text(
                j, i,
                f"{pct:+.1f}%",
                ha="center",
                va="center",
                fontsize=8,
                color="black"
            )

    ax.set_xticks(range(n_params))
    ax.set_xticklabels(param_labels, fontsize=11)

    ax.set_yticks(range(n_models))
    ax.set_yticklabels(labels, fontsize=10)

    ax.tick_params(length=0)

    cbar = fig.colorbar(im, ax=ax, pad=0.02, fraction=0.03)
    cbar.set_label("relative deviation ( (x - fid)/fid )", fontsize=9)
    cbar.ax.tick_params(labelsize=8)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.set_xticks(np.arange(n_params) - 0.5, minor=True)
    ax.set_yticks(np.arange(n_models) - 0.5, minor=True)
    ax.grid(which="minor", color="white", linewidth=1.5)
    ax.tick_params(which="minor", length=0)

    if title:
        ax.set_title(title, fontsize=12, pad=10)

    plt.tight_layout()
    plt.show()


plot_deviation_heatmap(build_hmf_paths("mor"),  HMFS_ALL, PARAMS_MOR,  title="HMF sensitivity — MOR")
plot_deviation_heatmap(build_hmf_paths("both"), HMFS_ALL, PARAMS_BOTH, title="HMF sensitivity — BOTH")

plot_deviation_heatmap(build_excl_paths("both"), ["exclude first bin", "baseline"], PARAMS_BOTH, title="Selection — BOTH")

plot_deviation_heatmap(build_cosmo_paths("both", lensing=False), COSMO_LABELS, PARAMS_BOTH, title="Delta Sigma Cosmology: full analysis")

plot_deviation_heatmap(lensing_paths, lensing_labels, PARAMS_BOTH, title="Delta Sigma Cosmology: lensing only")
```

```python
import pyccl as ccl
import numpy as np
```

```python
cosmo = ccl.Cosmology(
    Omega_c=0.2607,      # Cold dark matter density
    Omega_b=0.04897,     # Baryon density
    h=0.6766,            # Hubble parameter
    sigma8=0.8102,       # Matter fluctuation amplitude
    n_s=0.9665,          # Spectral index
)
z_edges = (0.5, 1)
z_edges_err = (0.5, 1.0, 1.2)
radius_center = np.atleast_1d(10.5)
z_bin_mean = (z_edges[1]+z_edges[0]) / 2.
a = 1.0 / (1.0 + z_bin_mean)
angular_center = radius_center / cosmo.angular_diameter_distance(a)  # radians
angular_center_arcmin = angular_center * 180.0 / np.pi * 60.0        # arcmin
radius_center_2 = np.radians(angular_center_arcmin) * cosmo.angular_diameter_distance(a) / 60
print(angular_center_arcmin, radius_center_2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
sacc_path = "/sps/lsst/groups/clusters/cl_pipeline_project/TXPipe_data/cosmodc2/outputs-full-10mpc-2026/cluster_sacc_catalog.sacc"
config_path = "/sps/lsst/users/ebarroso/CLPipeline/examples/cosmoDC2-full/cosmodc2_config_in2p3.yml"
set_params = {"mass_distribution_mu_p0": 3.35, "mass_distribution_mu1": 2.23 / np.log(10), "mass_distribution_mu2": 0.06, "mass_distribution_sigma0": 0.56, "mass_distribution_sigma1": 0.10 / np.log(10), "mass_distribution_sigma2": -0.05, "cluster_theory_cluster_concentration" : 1.}
# Get your recipes
cluster_counts_theory, cluster_counts_data, cluster_shear_profile_theory, cluster_shear_data = build_cluster_recipes_from_config(
    config_path, sacc_path, set_params = set_params
)

# Convert to numpy arrays if needed
counts_theory = np.array(cluster_counts_theory)
counts_data = np.array(cluster_counts_data)
shear_theory = np.array(cluster_shear_profile_theory)
shear_data = np.array(cluster_shear_data)

# Differences
diff_counts = counts_data - counts_theory
diff_shear = shear_data - shear_theory

# --- Plot cluster counts ---
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(counts_theory, label="Theory", marker='o')
plt.plot(counts_data, label="Data", marker='x')
plt.plot(diff_counts, label="Data - Theory", linestyle='--', color='gray')
plt.xlabel("Bin index")
plt.ylabel("Cluster counts")
plt.title("Cluster Counts: Theory vs Data")
plt.legend()
plt.grid(True)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot((counts_data - counts_theory) / counts_data, label="Rel_diff", marker='o')

plt.xlabel("Bin index")
plt.ylabel("Cluster counts")
plt.title("Cluster Counts: Theory vs Data")
plt.legend()
plt.grid(True)



# --- Plot shear profile ---
plt.subplot(1, 2, 2)
plt.plot(shear_theory, label="Theory", marker='o')
plt.plot(shear_data, label="Data", marker='x')
plt.plot(diff_shear, label="Data - Theory", linestyle='--', color='gray')
plt.xlabel("Bin index")
plt.ylabel("Shear")
plt.title("Cluster Shear Profile: Theory vs Data")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
